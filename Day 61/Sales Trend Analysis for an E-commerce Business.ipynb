Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Sample dataset

```python
data = {
    "Month": ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "Revenue": [12000,15000,17000,14000,18000,20000,22000,25000,24000,26000,30000,35000],
    "Customers": [200,230,250,210,260,280,300,330,320,350,400,450],
    "Category": ["Electronics","Fashion","Home & Kitchen","Sports","Books",
                 "Electronics","Fashion","Home & Kitchen","Sports","Books",
                 "Electronics","Fashion"],
    "CategorySales": [5000,4000,3000,2000,1000,6000,4500,3500,2200,1200,8000,7000],
    "Orders": [2,5,3,8,4,6,7,3,9,4,6,8],
    "TotalSpent": [200,600,350,1200,500,900,1100,400,1500,550,800,1200]
}

df = pd.DataFrame(data)
print(df.head())
```

Data Visualizations with Seaborn

(a) Monthly Revenue Trend

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Month", y="Revenue", data=df, marker="o", color="blue")
plt.title("Monthly Revenue Trend")
plt.ylabel("Revenue ($)")
plt.xlabel("Month")
plt.show()
```

(b) Customer Spending Behavior (Scatter Plot)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Orders", y="TotalSpent", data=df, hue="Category", s=100, palette="viridis")
plt.title("Customer Orders vs Total Spent")
plt.xlabel("Number of Orders")
plt.ylabel("Total Spent ($)")
plt.show()
```

(c) Product Category Sales (Bar Chart)

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="CategorySales", data=df, palette="pastel")
plt.title("Sales by Product Category")
plt.ylabel("Sales ($)")
plt.xlabel("Product Category")
plt.xticks(rotation=30)
plt.show()
```

### Insights You Can Draw

- Revenue increases steadily with strong growth in the last quarter.

- Customers with higher orders spend more, showing loyalty impacts revenue.

- Electronics & Fashion are top-performing categories, while Books contribute least.
