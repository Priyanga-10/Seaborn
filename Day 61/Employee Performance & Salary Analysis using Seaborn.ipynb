Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

Sample employee dataset

```python
data = {
    "ExperienceLevel": ["Junior", "Junior", "Mid", "Mid", "Mid", 
                        "Senior", "Senior", "Senior", "Junior", "Mid"],
    "Salary": [35000, 37000, 55000, 60000, 58000, 
               90000, 95000, 92000, 40000, 62000],
    "PerformanceRating": [3, 4, 2, 5, 3, 4, 5, 4, 3, 5],
    "Gender": ["Male", "Female", "Male", "Female", "Male", 
               "Female", "Male", "Female", "Male", "Female"]
}

df = pd.DataFrame(data)
print(df)
```

Displaying Summary Statistics

```python
print("Dataset Info:")
print(df.info())

print("\nDataset Description:")
print(df.describe())
```

Line Plot – Salary Growth Over Experience Levels

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=df, x="ExperienceLevel", y="Salary", marker="o")
plt.title("Salary Growth by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.show()
```

Scatter Plot – Performance vs Salary

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="PerformanceRating", y="Salary", hue="Gender", style="Gender", palette="coolwarm", s=100)
plt.title("Performance vs Salary")
plt.xlabel("Performance Rating")
plt.ylabel("Salary")
plt.show()
```

Bar Plot – Average Salary by Experience Level

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="ExperienceLevel", y="Salary", palette="pastel")
plt.title("Average Salary Based on Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")
plt.show()
```

Gender-Based Salary Comparison

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="Gender", y="Salary", palette="viridis")
plt.title("Gender-Based Salary Comparison")
plt.xlabel("Gender")
plt.ylabel("Average Salary")
plt.show()
```

Changing Color Palettes

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="ExperienceLevel", y="Salary", palette="coolwarm")
plt.title("Experimenting with Color Palette (coolwarm)")
plt.show()
```

Exporting Visualizations

```python
plt.figure(figsize=(6,4))
sns.lineplot(data=df, x="ExperienceLevel", y="Salary", marker="o")
plt.title("Export Example - Salary Growth by Experience")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary")

# Export to PNG & PDF
plt.savefig("salary_growth.png", dpi=300)
plt.savefig("salary_growth.pdf")
plt.show()
```

### Insights & Business Recommendations:

1. Salary Growth – Junior → Mid → Senior shows a clear jump; retention may depend on ensuring strong increments when moving up.

2. Performance vs Salary – High salaries are associated with higher ratings at senior levels → performance incentives seem effective.

3. Gender Gap – The given sample shows females at senior roles earn comparably, but monitoring is needed to avoid hidden disparities.

4. Experience-Based Pay – Juniors show close salary clustering → could consider performance-based differentiation.
