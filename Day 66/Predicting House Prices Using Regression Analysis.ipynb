###  Importing Required Libraries

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score 
```

### Load the Dataset

```python
data = {
    "SquareFootage": [1000, 1500, 2000, 2500, 3000, 3500],
    "Rooms": [3, 4, 3, 5, 4, 6],
    "Price": [200000, 250000, 270000, 350000, 400000, 450000],
    "Year": [2015, 2016, 2017, 2018, 2019, 2020] 
}

df = pd.DataFrame(data)
print(df)
```

###  Data Preprocessing

```python
print(df.info())
print(df.isnull().sum())
```

### Exploratory Data Analysis (EDA)

```python
sns.scatterplot(x="SquareFootage", y="Price", data=df)
plt.title("House Price vs Square Footage")
plt.show()
```

### Create Regression Plot

```python
sns.regplot(x="SquareFootage", y="Price", data=df, color="blue")
plt.title("Regression Plot: Price vs Square Footage")
plt.show()
```

### Customize Regression Line

```python
sns.regplot(x="SquareFootage", y="Price", data=df,  line_kws={'color': 'red', 'linewidth': 2}) 
plt.title("Customized Regression Line")
plt.show()
```

### Residual Plot

```python
sns.residplot(x="SquareFootage", y="Price", data=df, scatter_kws={'color': 'blue'})
plt.title("Residual Plot")
plt.show()
```

### Modeling (Linear Regression)

```python
X = df[["SquareFootage"]]  # Independent variable
y = df["Price"]            # Dependent variable

model = LinearRegression()
model.fit(X, y)
```

### Predictions

```python
df["PredictedPrice"] = model.predict(X)
print(df[["SquareFootage", "Price", "PredictedPrice"]])
```

### Model Evaluation

```python
r2 = r2_score(y, df["PredictedPrice"])
mse = mean_squared_error(y, df["PredictedPrice"])

print("R-squared:", r2)
print("Mean Squared Error:", mse)
```
