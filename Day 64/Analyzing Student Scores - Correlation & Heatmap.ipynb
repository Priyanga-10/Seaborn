### Import Libraries

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Create a DataFrame with Student Scores

```python
data = {
    'Math': [88, 92, 80, 89, 78], 
    'English': [85, 87, 82, 91, 76], 
    'Science': [90, 92, 88, 95, 80] 
} 
df = pd.DataFrame(data)
print(df)
```

### Create the Correlation Matrix using pandas

```python
correlation_matrix = df.corr() 
print("Correlation Matrix:\n", correlation_matrix)
```

### Generate a Heatmap using seaborn

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Correlation Matrix of Student Scores') 
plt.show()
```

###  Customize the Heatmap

```python
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linewidths=1, linecolor='black', fmt='.2f') 
plt.title('Customized Correlation Matrix of Student Scores') 
plt.show() 
```

### Handle Missing Data (imputing with mean if necessary) 

```python
df['Math'][2] = None  # Simulating a missing value in the Math column 
df.fillna(df.mean(), inplace=True) 
```

###  Recompute and plot the heatmap after handling missing data 

```python
correlation_matrix = df.corr() 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Correlation Matrix After Handling Missing Data') 
plt.show() 
```
