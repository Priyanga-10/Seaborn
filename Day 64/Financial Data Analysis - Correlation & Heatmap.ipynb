### Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### Create Dataset

```python
data = {
    'Company': ['A', 'B', 'C', 'D', 'E'],
    'StockPrice': [120, 150, 100, 130, 140],
    'Revenue': [2000, 2500, 1500, 2300, 2100],
    'Expenses': [1500, 2000, None, 1900, 1600],
    'ProfitMargin': [0.25, 0.20, 0.15, 0.22, 0.23]
}

df = pd.DataFrame(data)
print(df)
```

### Check for Missing Values

```python
print(df.isnull().sum())
```

### Handle Missing Values

```python
df = df.fillna(df.mean(numeric_only=True))
print(df)
```

### Calculate Correlation Matrix

```python
corr_matrix = df.drop('Company', axis=1).corr()
print(corr_matrix)
```

### Display the Correlation Matrix (Readable Format)

```python
pd.set_option("display.precision", 2)
print(corr_matrix)
```

### Create Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap - Financial Data", fontsize=14)
plt.show()
```

### Customize Heatmap Colors

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="viridis")
plt.title("Customized Correlation Heatmap", fontsize=14)
plt.show()
```

### Add Annotations with Formatting

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Heatmap with Annotations", fontsize=14)
plt.show()
```

### Adjust Line Widths & Borders

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=1, linecolor='black')
plt.title("Heatmap with Borders", fontsize=14)
plt.show()
```

### Handle Missing Data in Correlation Matrix

```python
mask = corr_matrix.isnull()
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap="coolwarm")
```

### Use a Custom Color Map

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="RdYlGn", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```

### Insights from the Heatmap

- Revenue & StockPrice  strong positive correlation (higher revenue, higher stock price).

- Expenses & ProfitMargin  strong negative correlation (higher expenses reduce margin).

- Revenue & ProfitMargin  moderate positive correlation (growing revenue improves margins).
