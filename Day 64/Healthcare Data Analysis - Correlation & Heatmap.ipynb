### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data Creation

```python
data = {
    "Age": [25, 32, 45, 50, 36, 60, 55, 29, 40, 70],
    "BloodPressure": [120, 130, 140, 135, 125, 150, 145, 128, 138, 160],
    "Cholesterol": [180, 190, 210, 195, 185, 220, 215, 200, 205, 230],
    "BMI": [22.5, 24.0, 27.5, 28.0, 25.0, 30.5, 29.0, 23.0, 26.5, 31.0],
    "Glucose": [90, 95, 110, 105, 100, 120, 115, 98, 108, 125],
    "HeartRate": [80, 78, 76, 74, 79, 70, 72, 82, 75, 68]
}
df = pd.DataFrame(data)
print(df.head())
```

### Data Preprocessing

```python
# Check data types
print("Data Types:\n", df.dtypes)

# Summary statistics
print("\nSummary Statistics:\n", df.describe())
```

### Check for Missing Values

```python
print("\nMissing Values:\n", df.isnull().sum())
```

### Calculate Correlation Matrix

```python
corr_matrix = df.corr()
print("\nCorrelation Matrix:\n", corr_matrix)
```

### Visualize Correlation Matrix

```python
print("\nCorrelation Matrix (Rounded):\n", corr_matrix.round(2))
```

### Create a Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Health Metrics")
plt.show()
```

### Customize Heatmap Color Map & Add Annotations

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,
    annot=True,       
    fmt=".2f",
    cmap="viridis", 
    linewidths=0.5,   
    linecolor="black"
)
plt.title("Customized Correlation Heatmap", fontsize=14)
plt.show()
```

### Create Custom Color Map

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix, 
    annot=True, 
    cmap="RdYlGn",  
    linewidths=0.3, 
    linecolor="grey"
)
plt.title("Heatmap with RdYlGn Color Map")
plt.show()
```

### Draw Conclusions

- Age & Blood Pressure  Strong positive correlation (0.97)  older patients tend to have higher BP.

- Age & Cholesterol  Positive correlation (0.96).

- BMI & Glucose  Moderate positive correlation (~0.91).

- Heart Rate & Age  Strong negative correlation (~ -0.97)  heart rate decreases as age increases.
