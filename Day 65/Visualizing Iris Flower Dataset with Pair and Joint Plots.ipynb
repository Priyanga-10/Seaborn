### Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Load the Iris Dataset

```python
iris = sns.load_dataset('iris')
print(iris.head())
```

### Explore the Dataset

```python
# Check dataset info
iris.info()

# Summary statistics
print(iris.describe())

# Check for missing values
print(iris.isnull().sum())
```

### Create a Pair Plot

```python
sns.pairplot(iris, kind="scatter")
plt.show()
```

### Add Color Differentiation Using hue

```python
sns.pairplot(iris, hue='species')
plt.show()
```

### Change Plot Kind to KDE

```python
sns.pairplot(iris, hue='species', kind='kde')
plt.show()
```

### Create a Joint Plot for Petal Length vs Petal Width

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='scatter')
plt.show()
```

### Hex Plot for Joint Plot

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='hex')
plt.show()
```

### KDE in Joint Plot

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='kde')
plt.show()
```

### Add a Regression Line

```python
sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='reg')
plt.show()
```

### Explore Sepal Width vs Sepal Length

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter')
plt.show()
```

### Visualize Petal Length vs Sepal Length

```python
sns.jointplot(x='sepal_length', y='petal_length', data=iris, kind='scatter')
plt.show()
```

### Export the Plots

```python
# Example for pair plot
pair_plot = sns.pairplot(iris, hue='species')
pair_plot.savefig("iris_pairplot.png") 

# Example for joint plot
joint_plot = sns.jointplot(x='petal_length', y='petal_width', data=iris, kind='reg')
joint_plot.savefig("iris_jointplot.png")
```
