### Import Libraries

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

### Load the Iris Dataset

```python
iris = sns.load_dataset('iris') 
print(iris.head())
```

### Create a Pair Plot

```python
sns.pairplot(iris, hue='species', palette='Set2') 
plt.show() 
```

### Customize Pair Plot with Hue 

```python
sns.pairplot(iris, hue='species', palette='Set1', kind='scatter') 
plt.show()
```

### Change Plot Kind in Pair Plot (using Regression Line)

```python
sns.pairplot(iris, hue='species', kind='reg', height=2.5) 
plt.show()
```

###  Adjust Plot Size

```python
sns.pairplot(iris, hue='species', height=3) 
plt.show() 
```

### Create a Joint Plot (Scatter + Histograms)

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter') 
plt.show()
```

### Customize Joint Plot with Different Kinds

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='hex', color='purple') 
plt.show() 
```

### Customize Joint Plot with Hue 

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde', hue='species') 
plt.show() 
```

###  Add Regression Line in Joint Plot 

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='reg', color='green') 
plt.show()
```

### Save the plots

```python
sns.pairplot(iris, hue='species') 
plt.savefig('iris_pairplot.png') 
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='scatter') 
plt.savefig('iris_jointplot.png')
```
