### Import Libraries

```python
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

###  Load the Wine Quality Dataset

```python
data = {
    "alcohol": [9.4, 10.0, 10.5, 11.2, 9.8, 12.0, 12.5, 10.8, 11.5, 9.7],
    "volatile_acidity": [0.70, 0.65, 0.58, 0.56, 0.60, 0.55, 0.50, 0.66, 0.62, 0.59],
    "citric_acid": [0.00, 0.04, 0.20, 0.24, 0.15, 0.30, 0.35, 0.25, 0.28, 0.18],
    "residual_sugar": [1.9, 2.1, 2.3, 2.6, 2.2, 2.8, 3.0, 2.5, 2.4, 2.0],
    "pH": [3.51, 3.40, 3.39, 3.36, 3.41, 3.30, 3.28, 3.42, 3.35, 3.38],
    "quality": [5, 5, 6, 6, 5, 7, 7, 6, 7, 5]
}

df = pd.DataFrame(data)
print("Dataset preview:")
print(df.head())
```

###  Create a Pair Plot for All Variables

```python
sns.pairplot(df, hue='quality', palette='viridis') 
plt.show()
```

### Customize Pair Plot with Hue for Quality

```python
sns.pairplot(df, hue='quality', palette='coolwarm', kind='scatter') 
plt.show()
```

### Change Plot Kind in Pair Plot (using KDE) 

```python
sns.pairplot(df, hue='quality', kind='kde', height=2.5) 
plt.show()
```

### Adjust Plot Size 

```python
sns.pairplot(df, hue='quality', height=3) 
plt.show()
```

### Create a Joint Plot between Alcohol and Quality

```python
sns.jointplot(x='alcohol', y='quality', data = df, kind='scatter') 
plt.show()
```

### Change the Joint Plot to Hex Plot

```python
sns.jointplot(x='alcohol', y='quality', data = df, kind='hex', color='green') 
plt.show()
```

###  Use KDE in the Joint Plot

```python
sns.jointplot(x='alcohol', y='quality', data = df, kind='kde', color='orange') 
plt.show() 
```

### Add Regression Line to Joint Plot

```python
sns.jointplot(x='alcohol', y='quality', data = df, kind='reg', color='purple') 
plt.show()
```

### Visualize the Relationship Between pH and Quality 

```python
sns.jointplot(x='pH', y='quality', data=df, kind='scatter') 
plt.show()
```

### Explore the Relationship Between Residual Sugar and Quality 

```python
sns.pairplot(df, hue='quality', vars=['residual_sugar', 'quality', 'alcohol']) 
plt.show()
```

###  Handle Missing Data

```python
df = df.dropna()  
```

### Save and Export the Plots

```python
sns.pairplot(df, hue='quality', palette='coolwarm') 
plt.savefig('wine_pairplot.png') 
 
sns.jointplot(x='alcohol', y='quality', data=df, kind='scatter') 
plt.savefig('wine_jointplot.png') 
```
