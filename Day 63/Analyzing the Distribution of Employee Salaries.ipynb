### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load the employee salary dataset

```python
df = pd.read_csv("employee_salaries.csv")
print(df.head())
```

### Check for missing values

```python
print("Missing Values:\n", df.isnull().sum())
```

### Basic histogram of Salary

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Salary'], bins=10, kde=False, color="skyblue", edgecolor="black")
plt.xlabel("Salary in USD")
plt.ylabel("Frequency")
plt.title("Distribution of Employee Salaries")
plt.show()
```

### Histogram with KDE overlay

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Salary'], bins=10, kde=True, color="lightgreen", edgecolor="black")
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Employee Salary Distribution with KDE")
plt.show()
```

### Customize bin sizes

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Salary'], bins=15, kde=True, color="orange", edgecolor="black")
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Salary Distribution (Custom Bins)")
plt.show()
```

### Salary distribution by Department (using hue)

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Salary", hue="Department", bins=10, kde=True)
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Salary Distribution by Department")
plt.show()
```

### Distplot (displot) combining histogram + KDE

```python
sns.displot(df, x="Salary", kde=True, bins=10, height=5)
plt.xlabel("Salary in USD")
plt.title("Combined Histogram & KDE Plot of Salaries")
plt.show()
```

### Multiple KDE lines by Department

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x="Salary", hue="Department", linewidth=2)
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Salary KDE Comparison Across Departments")
plt.show()
```

### Custom colors with palette

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Salary", hue="Department", bins=10, kde=True, palette="Set2", edgecolor="black")
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Salary Distribution by Department (Custom Colors)")
plt.show()
```

### Save final visualization

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x="Salary", hue="Department", fill=True, palette="husl")
plt.xlabel("Salary in USD")
plt.ylabel("Density")
plt.title("Final Salary Distribution Comparison by Department")
plt.savefig("salary_distribution.png", dpi=300)
plt.show()
```
