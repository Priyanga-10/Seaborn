### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Sample dataset

```python
data = {
    "Math":    [88, 72, 95, 66, 54, 77, 81, 69, 92, 58],
    "Science": [75, 68, 89, 70, 60, 80, 85, 72, 90, 65],
    "English": [82, 79, 91, 73, 59, 85, 88, 76, 94, 62],
    "Gender":  ["Male", "Female", "Male", "Female", "Male",
                "Female", "Male", "Female", "Male", "Female"]
}

df = pd.DataFrame(data)
print("Dataset:\n", df.head())
```

### Check missing values

```python
print("\nMissing values:\n", df.isnull().sum())
```

###  Histogram for each subject

```python
subjects = ["Math", "Science", "English"]
for subj in subjects:
    sns.histplot(df[subj], bins=5, kde=False, color="skyblue", edgecolor="black")
    plt.title(f"{subj} Score Distribution")
    plt.xlabel("Score")
    plt.ylabel("Count")
    plt.show()
```

### Histogram + KDE for each subject

```python
subjects = ["Math", "Science", "English"]
for subj in subjects:
    sns.histplot(df[subj], bins=5, kde=True, color="green", edgecolor="black")
    plt.title(f"{subj} Score Distribution")
    plt.xlabel("Score")
    plt.ylabel("Count")
    plt.show()
```

### Histogram + KDE with gender hue

```python
for subj in subjects:
    sns.displot(df, x=subj, hue="Gender", kind="hist", kde=True, bins=5, palette="Set2", height=4, aspect=1.3)
    plt.title(f"{subj} Scores by Gender")
    plt.show()
```

### Multiple KDE lines (by Gender)

```python
for subj in subjects:
    sns.kdeplot(data=df, x=subj, hue="Gender", fill=False, palette="coolwarm", linewidth=2)
    plt.title(f"{subj} KDE by Gender")
    plt.xlabel("Score")
    plt.ylabel("Density")
    plt.show()
```

### Save one example plot

```python
sns.displot(df, x="Math", hue="Gender", kind="hist", kde=True, bins=5, palette="muted")
plt.title("Math Scores by Gender")
plt.savefig("math_scores.png")
plt.show()
```

### Final combined plot

```python
plt.figure(figsize=(12, 4))
for i, subj in enumerate(subjects, 1):
    plt.subplot(1, 3, i)
    sns.histplot(df, x=subj, hue="Gender", bins=5, kde=True, palette="muted")
    plt.title(f"{subj}")
    plt.xlabel("Score")
    plt.ylabel("Count")
plt.tight_layout()
plt.show()
```
