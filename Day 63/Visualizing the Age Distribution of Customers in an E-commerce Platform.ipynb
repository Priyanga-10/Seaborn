### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Customers Data

```python
df = pd.read_csv("customers.csv")
print(df.head())
```

### Handle Missing Values & Outliers

```python
print(df.isnull().sum())
```

### Histogram of Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=15, kde=False)
plt.xlabel("Customer Age")
plt.ylabel("Count")
plt.title("Age Distribution of Customers")
plt.show()
```

### Histogram + KDE Plot

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=15, kde=True)
plt.xlabel("Customer Age")
plt.ylabel("Density")
plt.title("Age Distribution with KDE Curve")
plt.show()
```

### Adjust Number of Bins

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True)
plt.xlabel("Customer Age")
plt.ylabel("Density")
plt.title("Age Distribution (Adjusted Bins)")
plt.show()
```

### Age Distribution by Gender (Hue Parameter)

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Age", hue="Gender", bins=20, kde=True, palette="Set2")
plt.xlabel("Customer Age")
plt.ylabel("Density")
plt.title("Age Distribution by Gender")
plt.show()
```

### Combine the Histogram and KDE plot with a faceted grid

```python
sns.displot(data=df, x="Age", hue="Gender", kind="hist", kde=True, bins=20, height=5, aspect=1.3, palette="Set2")
plt.xlabel("Customer Age")
plt.ylabel("Density")
plt.title("Faceted Age Distribution by Gender")
plt.show()
```

### Multiple KDE Lines for Gender

```python
plt.figure(figsize=(8,5))
sns.kdeplot(data=df, x="Age", hue="Gender", fill=False, palette="Set1")
plt.xlabel("Customer Age")
plt.ylabel("Density")
plt.title("KDE Plot of Age Distribution by Gender")
plt.show()
```

### Custom Labels, Title, and Color Palette

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Age", hue="Gender", bins=20, kde=True, palette="coolwarm")
plt.xlabel("Customer Age", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.title("Customer Age Distribution by Gender", fontsize=14, fontweight="bold")
plt.show()
```

### Save the Final Plot

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x="Age", hue="Gender", bins=20, kde=True, palette="coolwarm")
plt.xlabel("Customer Age")
plt.ylabel("Number of Customers")
plt.title("Final Age Distribution by Gender")
plt.savefig("age_distribution_by_gender.png", dpi=300)
plt.show()
```
