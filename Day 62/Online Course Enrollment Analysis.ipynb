#### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Sample dataset

```python
data = {
    "Course": [
        "Python for Data Science", "Web Development Bootcamp", "Business Analytics",
        "Machine Learning A-Z", "JavaScript Essentials", "Digital Marketing",
        "Deep Learning Mastery", "React Web Development", "Entrepreneurship 101"
    ],
    "Category": [
        "Data Science", "Web Development", "Business",
        "Data Science", "Web Development", "Business",
        "Data Science", "Web Development", "Business"
    ],
    "Enrollments": [350, 420, 280, 500, 390, 310, 450, 370, 260],
    "Rating": [4.6, 4.5, 4.2, 4.7, 4.3, 4.1, 4.8, 4.4, 4.0],
    "Difficulty": [
        "Beginner", "Beginner", "Intermediate",
        "Advanced", "Beginner", "Intermediate",
        "Advanced", "Intermediate", "Beginner"
    ]
}

df = pd.DataFrame(data)
print(df)
```

#### Set Seaborn theme

```python
sns.set_theme(style="whitegrid", palette="Set2")
```

#### 1. Bar Plot – Average Enrollment per Category

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="Enrollments", data=df)
plt.title("Average Student Enrollments per Course Category")
plt.show()
```

####  2. Count Plot – Number of Courses per Category

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Category", data=df)
plt.title("Number of Courses per Category")
plt.show()
```

#### 3. Box Plot – Ratings by Course Category

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Category", y="Rating", data=df)
plt.title("Variation in Student Ratings Across Categories")
plt.show()
```

#### 4. Violin Plot – Enrollment Distribution by Difficulty Level

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="Difficulty", y="Enrollments", data=df)
plt.title("Distribution of Enrollments by Difficulty Level")
plt.show()
```

#### 5. Swarm Plot – Individual Course Enrollments by Category

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x="Category", y="Enrollments", data=df, hue="Course", palette="muted", size=8)
plt.title("Individual Course Enrollments by Category (with Course)")
plt.legend(title="Course")
plt.show()
```
