#### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Create Sample Data

```python
data = {
    "Product_Category": ["Electronics", "Electronics", "Groceries", "Groceries", "Clothing",
                         "Clothing", "Electronics", "Groceries", "Clothing", "Electronics"],
    "Revenue": [1200, 900, 200, 250, 400, 350, 1500, 300, 450, 800],
    "Price": [600, 450, 50, 40, 200, 180, 750, 60, 220, 500],
    "Discount": [10, 5, 2, 3, 15, 20, 8, 1, 10, 5],
    "Units_Sold": [2, 2, 4, 5, 2, 2, 2, 5, 2, 2]
}

df = pd.DataFrame(data)
print(df)
```

#### Set Seaborn theme

```python
sns.set_theme(style="darkgrid", palette="Paired")
```

#### 1. Bar Plot â€“ Average Revenue by Product Category

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Product_Category", y="Revenue", data=df, ci=None, palette="pastel")
plt.title("Average Revenue by Product Category")
plt.show()
```

#### 2. Count Plot â€“ Number of Purchases per Category

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Product_Category", data=df, palette="Set2")
plt.title("Number of Purchases per Product Category")
plt.show()
```

#### 3. Box Plot â€“ Product Price Variations

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Product_Category", y="Price", data=df, palette="Set3")
plt.title("Price Variation across Categories")
plt.show()
```

#### 4. Violin Plot â€“ Distribution of Discounts

```python
plt.figure(figsize=(7,5))
sns.violinplot(x="Product_Category", y="Discount", data=df, palette="muted")
plt.title("Discount Distribution by Product Category")
plt.show()
```

#### 5. Swarm Plot â€“ Individual Product Sales

```python
plt.figure(figsize=(7,5))
sns.swarmplot(x="Product_Category", y="Units_Sold", data=df, palette="viridis", size=7)
plt.title("Individual Product Sales by Category")
plt.show()
```
